```r
library(ggplot2)
library(data.table)
theme_set(theme_bw())
```

```r
strats = fread('stratified_genome_metadata.tsv')
setnames(strats, 'genome_ID', 'genome')
```

```r
tools = c('singlem', 'metaphlan', 'motus', 'kracken', 'sourmash')
```

```r
d1 = data.table(expand.grid(tool=tools, genome=strats$genome))
d1[1:3]
```

```r
# output_singlem/opal/GCA_022615805.1_genomic.opal_report
d2 = d1[, fread(paste('output_',tool,sep='','/opal/',genome,'.opal_report')), by=list(tool,genome)]
d2[1:3]
```

```r
dim(d2)
```

```r
setnames(d2, c('tool','genome','blah','rank','metric','sample','value'))
d3 = d2[blah != 'Gold standard'][,.(tool, genome, rank, metric, value)]
d3[1:3]
dim(d3)
```

```r
bc = d3[metric=='Bray-Curtis distance']
dim(bc)
bc[1:10]
```

```r
bc2 = merge(bc, strats, by='genome')
bc2[1:10]
```

```r
orders = c('kingdom','phylum','class','order','family','genus','species')
novel_orders = gsub('^','novel ',orders)
bc2[, rank := factor(rank, levels=orders)]
bc2[, novelty_category := factor(novelty_category, levels=orders)]
```

```r
bc2[, kingdom := gsub('d__', 'novel ', gsub(';.*','',taxonomy))]
bc2[, novelty_category_novel := factor(gsub('^','novel ',novelty_category), levels=novel_orders)]
bc2[1:3]
```

```r
qplot(data=bc2[rank=='kingdom'], tool, 1-value, geom='boxplot', main='Detection of novel lineages', colour=tool, ylab='Bray-Curtis similarity')+facet_grid(kingdom~novelty_category_novel)+theme(axis.text.x = element_text(angle = 45, hjust = 1))+theme(legend.position='none')
```

```r
# What about when the rank equals the novelty category?
qplot(data=bc2[rank==orders[as.numeric(novelty_category)-1]], main='Accuracy classifying novel lineages at highest possible resolution', ylab='Bray-Curtis similarity', tool, colour=tool, 1-value, geom='boxplot')+facet_grid(kingdom~novelty_category_novel)+theme(axis.text.x = element_text(angle = 45, hjust = 1))+labs(subtitle='e.g. Phylum level resolution for novel classes')+theme(legend.position='none')
```

```r
# What about when the rank equals one level above the novelty category?
qplot(data=bc2[novelty_category != 'phylum'][rank==orders[as.numeric(novelty_category)-2]], main='Accuracy of classification of novel lineages at 2nd highest possible resolution',  colour=tool, ylab='Bray-Curtis similarity', tool, 1-value, geom='boxplot')+facet_grid(kingdom~novelty_category_novel)+theme(axis.text.x = element_text(angle = 45, hjust = 1))+labs(subtitle='e.g. Phylum level resolution for novel orders')+theme(legend.position='none')
```
